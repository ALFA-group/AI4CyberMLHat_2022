```python
import pandas as pd

cwe_table = pd.read_csv("top_25_cwes.csv")
```

```python
cwe_table
```

```python
cwe_table.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
cwe_id = 732
cwe_id_info = cwe_table.loc[cwe_table["cwe id"] == cwe_id].iloc[0]
```

```python
cwe_id_info
```

```python
import networkx as nx

G = nx.Graph()
```

```python
G.add_node(cwe_id, label=cwe_id_info["name"], color="red", title="CWE")
G.add_node(
    "CWE mitigations",
    label=int(cwe_id_info["cwe mitigations"]),
    shape="diamond",
    color="red",
    title="CWE mitigations",
)
G.add_node(
    "CWE detections",
    label=int(cwe_id_info["cwe detections"]),
    shape="square",
    color="red",
    title="CWE detections",
)
G.add_node("AP", label=int(cwe_id_info["aps"]), color="green", title="Attack Patterns")
G.add_node(
    "AP mitigations",
    label=int(cwe_id_info["ap mitigations"]),
    shape="diamond",
    color="green",
    title="AP mitigations",
)
G.add_node(
    "AP detections",
    label=int(cwe_id_info["ap detections"]),
    shape="square",
    color="green",
    title="AP detections",
)
G.add_node(
    "Techniques",
    label=int(cwe_id_info["techniques"]),
    color="orange",
    title="Techniques",
)
G.add_node(
    "Technique mitigations",
    label=int(cwe_id_info["technique mitigations"]),
    shape="diamond",
    color="orange",
    title="Technique mitigations",
)
G.add_node(
    "Technique detections",
    label=int(cwe_id_info["technique detections"]),
    shape="square",
    color="orange",
    title="Technique detections",
)
G.add_node("Tactics", label=int(cwe_id_info["tactics"]), color="blue", title="Tactics")


G.add_edge(cwe_id, "CWE mitigations", color="black")
G.add_edge(cwe_id, "CWE detections", color="black")
G.add_edge(cwe_id, "AP", color="black")
G.add_edge("AP", "AP mitigations", color="black")
G.add_edge("AP", "AP detections", color="black")
G.add_edge("AP", "Techniques", color="black")
G.add_edge("Techniques", "Technique mitigations", color="black")
G.add_edge("Techniques", "Technique detections", color="black")
G.add_edge("Techniques", "Tactics", color="black")
```

```python
# Save graph
file_path = "demo/top_25_cwe_graph_networkx.pkl"
nx.write_gpickle(G, file_path)
```

```python
from pyvis.network import Network

nt = Network("600px", "600px", notebook=True)
nt.from_nx(G)
nt.show("top_25_cwe.html")
```
